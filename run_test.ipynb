```python
import backtest.run

mid, sigma, events = backtest.run.load_artifacts()
```

```python
import yaml

with open("config.yml", "r") as file:
    config = yaml.safe_load(file)

TEST_START = config['test_start']

test_results = backtest.run.run(mid, sigma, events, start = TEST_START, end = None)
```

```python
print((test_results['ask_quote'] - test_results['bid_quote']).mean() / 2) # half spread
```

```python
import backtest.symmetric_strategy

symmetric_results = backtest.run.run(mid, sigma, events,
                                     strategy_cls = backtest.symmetric_strategy.SymmetricStrategy,
                                     strategy_kwargs = {'half_spread': 5.02},
                                     start = TEST_START, end = None)
```

```python
test_results.to_parquet('binance/as_test_results')
symmetric_results.to_parquet('binance/symmetric_test_results')
```
