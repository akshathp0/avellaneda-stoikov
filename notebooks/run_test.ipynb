```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
```

```python
import backtest.run

mid, sigma, events = backtest.run.load_artifacts()
```

```python
from config import TEST_START, BINANCE_PATH

test_results = backtest.run.run(mid, sigma, events, start = TEST_START, end = None)
```

```python
half_spread = (test_results['ask_quote'] - test_results['bid_quote']).mean() / 2
```

```python
import backtest.symmetric_strategy

symmetric_results = backtest.run.run(mid, sigma, events,
                                     strategy_cls = backtest.symmetric_strategy.SymmetricStrategy,
                                     strategy_kwargs = {'half_spread': half_spread},
                                     start = TEST_START, end = None)
```

```python
test_results.to_parquet(BINANCE_PATH / 'as_test_results.parquet')
symmetric_results.to_parquet(BINANCE_PATH / 'symmetric_test_results.parquet')
```
